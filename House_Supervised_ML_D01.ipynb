# Dealing with Missing data

```python
# import
from sklearn.impute import SimpleImputer

# initialize
my_imputer = SimpleImputer(strategy = 'constant', fill_value = 1)

# fit
my_imputer.fit(X_num_train)

# transform train 
X_num_imputed_train = my_imputer.transform(X_num_train)
```

```python
count_nan = X['LotFrontage'].isnull().sum()
print ('Count of NaN: ' + str(count_nan))
```

```python
# 1. import the model
from sklearn.tree import DecisionTreeClassifier

# 2. initialize the model (optional: give some parameters)
my_tree = DecisionTreeClassifier()

# 3. fit the model to the train data
my_tree.fit(X = X_num_imputed_train, 
            y = y_train)
```

```python
# make predictions
y_pred_tree_train = my_tree.predict(X_num_imputed_train)
```

```python
# Check accuracy on the train set
from sklearn.metrics import accuracy_score

accuracy_score(y_true = y_train,
               y_pred = y_pred_tree_train
              )
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# 2. initialize the model (optional: give some parameters)
my_tree1 = DecisionTreeClassifier(max_depth= 23, min_samples_split=30)
my_tree1.fit(X = X_num_imputed_train, 
            y = y_train)

plt.figure(figsize=(40, 20))

plot_tree(
     my_tree1,
     filled=True, 
     rounded=True, 
     class_names=['Not Expensive','Expensive'], 
     feature_names=X_num.columns
     );
```
